```python
%sql
REFRESH TABLE avireportscomercial.caidas;
select
TIPO_DE_CAIDA,
count(distinct(session)) as SESIONES
from avireportscomercial.caidas
where year = 2024 and month in (4,5,6)
group by TIPO_DE_CAIDA
order by SESIONES desc
```
